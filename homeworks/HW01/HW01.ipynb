```python
import numpy as np
```

```python
a = np.array([1,2,3,4]) #создаем массив из стандартного массива 
print(a)
```

```python
a = np.zeros(4) #создаем одномерный массив из нулей
print(a)
```

```python
a = np.ones((4, 2)) #создаем матрицу формы 4x2 из единиц
print(a)
```

```python
a = np.arange(5) #создаем список из чисел от 0 до 4
print(a)
print(a.shape) #форма
print(a.dtype) #тип данных внутри
```

```python
a = np.arange(20)
print(a[5]) #конкретный элемент
print(a[2:12]) #срез
```

```python
a = np.arange(10)
b = np.arange(5, 15)
print(a, b)
print(a + b) #складываем векторы
```

```python
a = np.arange(10, 31)
print(a.mean()) #среднее значение. (10 + 30)/ 2 = 20
```
